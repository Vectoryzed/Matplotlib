```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from matplotlib.lines import Line2D

%matplotlib inline
```

```python
data = pd.read_csv('Data/barchart-divorcerates.csv')
data.head()
```

```python
fig, ax1 = plt.subplots(figsize=(12,4))

data.plot.bar(ax=ax1)
ax1.legend(ncol=2, loc='upper center')
```

```python
t25 = ax1.yaxis.get_major_ticks()[5]
t25.label2On = True
t25.tick2On = True
t25.label1On = False
t25.tick1On = False
fig
```

```python
ax1.tick_params(axis='y', labelsize=15)
fig
```

```python
ax1.tick_params(axis='x', labelrotation=15)
fig
```

```python
from matplotlib.ticker import StrMethodFormatter

mft = StrMethodFormatter("{x:.2f} USD")
ax1.yaxis.set_major_formatter(mft)
fig
```

```python
from matplotlib.ticker import FixedLocator

mnl = FixedLocator([0, 1, 3, 4])
ax1.xaxis.set_major_locator(mnl)
fig
```
