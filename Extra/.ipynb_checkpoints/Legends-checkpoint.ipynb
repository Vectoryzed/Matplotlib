```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from matplotlib.lines import Line2D

%matplotlib inline
```

```python
data = pd.read_csv('Data/barchart-divorcerates.csv')
data.head()
```

```python
fig, ax1 = plt.subplots(figsize=(12,4))

data.plot.bar(ax=ax1)
ax1.legend(ncol=2, loc='upper center')
```

```python
ax1.legend_.legendPatch.set(boxstyle='round', facecolor='#e0f7fa')
fig
```

```python
ax1.legend_.set_bbox_to_anchor([0.5, -0.1])
fig
```

```python
ax1.legend_.texts[0].set_color('red')
fig
```

```python
ax1.legend_.legendHandles[0].set_color('green')
fig
```
