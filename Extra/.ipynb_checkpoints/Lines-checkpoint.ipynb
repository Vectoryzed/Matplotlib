```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sns

from matplotlib.lines import Line2D

%matplotlib inline
```

```python
data = pd.read_csv('Data/iris.csv')
data.head()
```

```python
sns.jointplot('Sepal length', 'Petal width', data=data, kind='reg', ci=None, space=0)
fig = plt.gcf()
```

```python
# Get the main axes 
main_ax=fig.axes[0]

# Access the line 
line = main_ax.lines[0]
# Change styles 
line.set(color='maroon', linestyle='dotted')

fig
```

```python
line = fig.axes[0].lines[0] 
line.set_clip_on(False)
fig.axes[0].set_xlim(4,8) 
fig.axes[0].set_zorder(2) 
fig.axes[1].set_zorder(1) 
fig.axes[2].set_zorder(0)

fig
```

```python
line = fig.axes[0].lines[0]

line.set(marker='o', markevery=5, markerfacecolor='yellow')

fig
```
