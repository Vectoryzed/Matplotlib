```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sns
from matplotlib.lines import Line2D

%matplotlib inline
```

```python
data = pd.read_csv('Data/iris.csv')
data.head()
```

```python
fig, ax = plt.subplots(figsize=(12,4))
ax.plot('Sepal length', data=data) 
ax.set_title('Sepal Length Chart', size=22) 
xmax_pos = data['Sepal length'].values.argmax() 
ymax=data['Sepal length'].values.max()

ax.axvline(x=xmax_pos, color='red', ls='dashed') 
ann = ax.annotate(str(ymax), xy=(xmax_pos, ymax), xytext=(xmax_pos+10, ymax-1), arrowprops=dict(arrowstyle='->'))
```

```python
ann.set_bbox(dict(boxstyle='round, rounding_size=0.8, pad=8.9', facecolor='y')) 
fig
```

```python
ann.set_bbox(dict(boxstyle='round, rounding_size=0.8, pad=0.3', facecolor='y')) 

arrow = ann.arrow_patch 
arrow.set_arrowstyle('<->') 
arrow.set_connectionstyle('angle, angleA=90') 
fig
```
